Let's get % of population with at least a bachelor's degree

- Bachelor's degree: `B15003_022E`
- Master's degree: `B15003_023E`
- Professional school degree: `B15003_024E`
- Doctorate degree: `B15003_025E`
- total: `B15003_001E`

$$Share\ of\ Pop.\ with\ at\ least\ a\ bachelor's\ degree = \frac{B15003\_022E + B15003\_023E + B15003\_024E + B15003\_025E}{B15003\_001E}$$

````python
my_vars = ["B15003_022E", "B15003_023E", "B15003_024E", "B15003_025E", "B15003_001E"]
````

Grab your geographies

````python
geography_of_interest = [("state", "06"), ("county","*")]
my_geo = censusdata.censusgeo(geo = geography_of_interest)
````

# Download data

````python
censusdata.download?
````

````python
data = censusdata.download(src = 'acs5', year = 2017, geo = my_geo, var = my_vars)
````

````python
data.head()
````

````python
rename_vars = {
    "B15003_022E": "Bachelors", 
    "B15003_023E": "Masters", 
    "B15003_024E": "Professional degree", 
    "B15003_025E": "Doctorates",
    "B15003_001E": "Total",
}
````

````python
# need to do this because rename returns dataframe
data = data.rename(mapper = rename_vars, axis = 1)
````

````python
data.head()
````

````python
data['share of bachelors+'] = (data['Bachelors'] + data['Masters'] + data['Professional degree'] + data['Doctorates']) / data['Total']
````

````python
data.head()
````

# .copy() vs subset

Right now, clean_data is "pointing" (is a `view`) to a subset of data so any changes you do to clean_data aren't _guaranteed_ to work. sometimes they do sometimes they don't. So you should create a .copy() of the data if you plan on modifying it after.

````python
clean_data['State'] = 'CA'
````

````python
clean_data = data[columns_im_interested_in].copy()
````

````python
clean_data['State'] = 'CA'
````

No warning
